## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%
CREATE TABLE "usuarios"(
id_usuario INTEGER GENERATED BY DEFAULT AS IDENTITY,
nombre VARCHAR(45) NOT NULL,
apellido VARCHAR(45) NOT NULL,
correo VARCHAR(45) NOT NULL,
contraseña VARCHAR(50) NOT NULL,
fecha_registro TIMESTAMP NOT NULL,
PRIMARY KEY (id)
);
%sql
```

### Tabla "Creadores"

```python
%
CREATE TABLE "creadores"(
id_creador INTEGER GENERATED BY DEFAULT AS IDENTITY,
nombre VARCHAR(45) NOT NULL,
apellido VARCHAR(45) NOT NULL,
popularidad INTEGRER,
estilo VARCHAR(45),
PRIMARY KEY (id_creador)
);
%sql
```

### Tabla "Tipos_NTF"

```python
%
CREATE TABLE "tipos_nft"(
id_tipo INTEGER GENERATED BY DEFAULT AS IDENTITY,
nombre VARCHAR(20) NOT NULL,
PRIMARY KEY (id_tipo)
);
%sql
```

### Tabla "NFT"

```python
%
CREATE TABLE "nft"(
id_nft INTEGER GENERATED BY DEFAULT AS IDENTITY,
nombre VARCHAR(45) NOT NULL,
id_creador INTEGER NOT NULL,
id_tipo INTEGER,
PRIMARY KEY (id_nft),
FOREIGN KEY (id_creador) REFERENCES "creadores" (id_creador),
FOREIGN KEY (id_tipo) REFERENCES "tipos_nft" (id_tipo)
);
%sql
```

### Tabla "Precio NFT"

```python
%
CREATE TABLE "precio_nft"(
id_nft INTEGER,
fecha TIMESTAMP NOT NULL,
precio INTEGER NOT NULL,
PRIMARY KEY (fecha),
FOREIGN KEY (id_nft) REFERENCES "nft" (id_nft)
);
%sql
```

### Tabla "Usuario tiene NFT"

```python
%
CREATE TABLE "usuario_tiene_nft"(
id_usuario INTEGER,
id_nft INTEGER,
FOREIGN KEY (id_usuario) REFERENCES "usuarios" (id_usuario),
FOREIGN KEY (id_nft) REFERENCES "nft" (id_nft)
);
%sql
```

### Consulta 1
**ID y Nombre del creador de un estilo X con mayor popularidad.**
###### Si hay más de uno, ponga solamente el primero

Ejemplo:

|Id |  Creador      |
|---|-----------|
| 12 |Levi Ackerman |

**Solución:** _Accedimos a la tabla de "creadores" en la cual rescatamos el id, nombre y apellido del creador que tenga los requerimientos solicitados como lo son el estilo y quien sea más popular, luego ordenamos y limitamos el output a solo 1._


```python
%
SELECT
	id_creador,nombre,apellido
FROM
	creadores
WHERE
	estilo = 'Arte 3D'
ORDER BY
	popularidad DESC
LIMIT 1;
%sql
```

### Consulta 2
**Cantidad de precios que ha tenido históricamente cada NTF.**

Ejemplo:


| NFT     | Cantidad de precios |
|--------------|----------|
| Minecraft Sphere |    3     |
| The Binary Girl     |    7     |
| ...          |   ...    |

**Solución:** _Aqui optamos por una solucion anidada devido a que nos parecio lo más optimo a realizar, procedimos a seleccionar el nombre y contarlo cuantas veces se repita con el comando y le cambiamos el nombre al contador solo para efectos practicos, luego en el anidamiento juntamos las tablas para optener solo los valores requeridos y agrupamos._

```python
%
SELECT
	nombre,COUNT(tablanu.nombre) AS "Cantidad de precios"
FROM(
SELECT * FROM nft INNER JOIN precio_nft ON nft.id_nft = precio_nft.id_nft
) AS tablanu

GROUP BY tablanu.nombre
%sql
```

### Consulta 3
**El nombre y tipo (nombre del tipo) del NFT históricamente más caro.**
###### Si hay más de uno, ponga solamente el primero
Ejemplo:

| NFT    | Tipo | 
|----------------|--------|
| Minecraft Sphere| Arte 3D |

**Solución:** _FALTA DETALLAR EL CODIGO Y SACAR EL NOMBRE ESTA CASI ESTA COSA GENTE._

```python
%
SELECT 
	nombre, precio
FROM nft LEFT JOIN precio_nft ON precio_nft.id_nft = nft.id_nft
WHERE
	precio > 0
ORDER BY
	precio DESC
LIMIT 1
%sql
```

### Consulta 4
**Nombre y apellido del usuario con más número de NFTs.**

Ejemplo:

| Nombre     | Apellido |
|----------------|--------|
| Levi | Ackerman |

**Solución:** _Tomamos los datos utiles de la union de las tablas "usuarios" y "usuario_tiene_nft", agrupamos y ordenamos descendiente._

```python
%
SELECT
	nombre AS "Nombre",apellido AS "Apellido"
FROM(
	SELECT
	 nombre,apellido,COUNT(nombre) AS "Cantidad NFT"
FROM
	usuarios RIGHT JOIN usuario_tiene_nft ON usuarios.id_usuario = usuario_tiene_nft.id_usuario
GROUP BY
	nombre,apellido,nombre
ORDER BY
	COUNT(nombre) DESC
LIMIT 1
) AS machina
%sql
```

### Consulta 5
**El NFT que más veces cambió de precio historicamente.**

Ejemplo:

|  ID  | NFT | 
|--------|-------|
| 11 |Minecraft Sphere | 

**Solución:** _ESTAN ORDENADOS SOLO FALTA HACER EL JOIN CON LA OTRA TABLA Y SACAR EL NOMBRE._

```python
%
SELECT
	id_nft AS "ID",nombre AS "Nombre"
FROM(
	SELECT
	id_nft,nombre,COUNT(id_nft)
FROM 
	precio_nft NATURAL JOIN nft
GROUP BY
	nombre,id_nft,id_nft
ORDER BY
	COUNT(id_nft) DESC
LIMIT 1
) AS deus
%sql
```

### Consulta 6
**Usuarios que NO tienen un NFT.**

Ejemplo

|Nombre | Apellido|
|--------|-------|
|Guillermo    | Diaz |
|Markus | Person |
|Rick   | Astley |
|... | ...| 

**Solución:** _Aqui lo que hicimos fue tomar el nombre y apellido de una consulta anidada la cual nos juntaba las tablas "usuarios" y "usuario_tiene_nft" en la cual mediante una condición sacamos a los usuarios que su id no estaba presente en la tabla "usuario_tiene_nft", es decir, eran NULL._

```python
%
SELECT
	nombre AS "Nombre",apellido AS "Apellido"
FROM(
	SELECT
	nombre,apellido,usuario_tiene_nft.id_usuario,usuario_tiene_nft.id_nft
	FROM
	usuarios LEFT JOIN usuario_tiene_nft ON usuario_tiene_nft.id_usuario = usuarios.id_usuario
WHERE
	usuario_tiene_nft.id_usuario IS NULL
) AS apoyo
%sql
```

### Consulta 8
**Los precios de todos los NFTs entre los años 2017 y 2019 inclusive.**

Ejemplo:

|NFT |Precio |
|----|---|
|Minecraft Sphere |  4000 |
|The Binary Girl | 3000| 
| Koalas in Love | 6000 |
|Minecraft Sphere |  1250 |

**Solución:** _Para esta consulta, mediante el SELECT tomamos el nombre y precio (Que es lo solicitado), luego juntamos las tablas "nft" y "precio_nft" para acceder a la fecha, luego con el comando 'BETWEEN' tomamos desde el 2017 hasta el 2020._

```python
%
SELECT
	nombre,precio
FROM 
	nft LEFT JOIN precio_nft ON nft.id_nft = precio_nft.id_nft
WHERE
	fecha BETWEEN '2016-12-31' AND '2020-01-01'
%sql
```
